### The Code

```python
def submit_treasuremap_batch(graceid, api_token, status, ralist, declist, times, depths, bands, instrumentid):
	#Bulk submission of treasure map pointings for e.g. Swope in the format e.g. # name ra dec filter exptime start_time_utc m3sigma 
    #Assumes we are using the AB magnitude system and are not savages. 
    
    batch = []

    for i in range(len(ralist)):
                
        p = gwtm_api.Pointing(
        ra=ralist[i],
        dec=declist[i],
        instrumentid = instrumentid,
        time = times[i][:21],
        status = status,
        depth = depths[i],
        depth_unit = 'ab_mag', 
        pos_angle=0,
        band = bands[i][:1]
        )
        
        print(ralist[i], times[i][:21])
    
        batch.append(p)
        
    print(batch[0].time)
    gwtm_api.Pointing.batch_post(pointings=batch, graceid=graceid, api_token=api_token)
    
    print(len(batch))
    print(len(ralist))
```

```python
def submit_treasuremap(graceid, api_token, status, ralist, declist, times, depths, bands, instrumentid):
	#Bulk submission of treasure map pointings for e.g. Swope in the format e.g. # name ra dec filter exptime start_time_utc m3sigma 
    #Assumes we are using the AB magnitude system and are not savages. 
    
    batch = []

    for i in range(len(ralist)):
            
        p = gwtm_api.Pointing(
        ra=ralist[i],
        dec=declist[i],
        instrumentid = instrumentid,
        time = times[i][:21],
        status = status,
        depth = depths[i],
        depth_unit = 'ab_mag', 
        pos_angle=0, 
        band = bands[i][:1]
        )
        
        print(ralist[i], times[i][:21])
        
        p.post(graceid=graceid, api_token=api_token)
        
```
